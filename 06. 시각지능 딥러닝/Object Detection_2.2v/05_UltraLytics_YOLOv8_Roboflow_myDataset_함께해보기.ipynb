### 1) 모델 구조만 빌려와서 사용해보기

* 모델 구조 선택하기

```python
model = YOLO(model='yolov8n.yaml', task='detect')
```

* 모델 학습하기

```python
model.train(data='/content/roboflowex2-1/data.yaml',
            epochs=150,
            patience=7,
            pretrained=False,
            verbose=True,
            seed=2023,
            )
```

* 예측해보기

```python
results = model.predict(source='/content/test01.jpg',
                        save=True,
                        line_width=2)
```

### 2) 모델 구조에 사전 학습 가중치도 가져와서 사용해보기

* 모델 구조 및 사전 학습 가중치 선택하기

```python
model2 = YOLO(model='yolov8n.pt', task='detect')
```

* 모델 학습하기

```python
model2.train(data='/content/roboflowex2-1/data.yaml',
            epochs=150,
            patience=10,
            pretrained=True,
            verbose=True,
            seed=2023,
            )
```

* 예측해보기

```python
results = model2.predict(source='/content/test01.jpg',
                        save=True,
                        line_width=2)
```
