```python
import torch

x_train = torch.FloatTensor([[0,0],[0,1],[1,0],[1,1]])
y_train = torch.FloatTensor([[0],[1],[1],[0]])
W_h = torch.randn([2,3], requires_grad=True)
b_h = torch.randn([3], requires_grad=True)
W_o = torch.randn([3,1], requires_grad=True)
b_o = torch.randn([1], requires_grad=True)

optimizer = torch.optim.SGD([W_h,W_o, b_h, b_o], lr=0.01)

# 딥러닝의 1단계: 모델을 만듦
def H(x):
    return torch.sigmoid(torch.matmul(x, W_h) + b_h) # H(x) = sigmoid(Wx + b)

# 딥러닝의 2단계: 학습을 한다. epochs=2000
for step in range(2000):
    cost = -torch.mean(y_train*torch.log(H(x_train)) + (1-y_train)*torch.log(1-H(x_train))) # cost = 평균{(H(x) - y)의 제곱}
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

# 딥러닝 3단계: 추론/테스트을 한다
# x_test = torch.FloatTensor([[4,4]])
print(H(x_train))

```
