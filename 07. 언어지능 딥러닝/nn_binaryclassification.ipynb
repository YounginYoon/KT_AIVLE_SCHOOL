```python
import torch

x_train = torch.FloatTensor([[1,2],[2,3],[3,4],[4,4],[5,3],[6,2]])
y_train = torch.FloatTensor([[0],[0],[0],[1],[1],[1]])
W = torch.randn([2,1], requires_grad=True)
b = torch.randn([1], requires_grad=True)

print(x_train.shape)
optimizer = torch.optim.SGD([W,b], lr=0.01)
print(W.shape)
# 딥러닝의 1단계: 모델을 만듦
def H(x):
    return torch.sigmoid(torch.matmul(x, W) + b) # H(x) = Wx + b

# 딥러닝의 2단계: 학습을 한다. epochs=2000
for step in range(200):
    cost = -torch.mean(y_train*torch.log(H(x_train)) + (1-y_train)*torch.log(1-H(x_train))) # cost = 평균{(H(x) - y)의 제곱}
    optimizer.zero_grad()
    cost.backward()
    optimizer.step()

# 딥러닝 3단계: 추론/테스트을 한다
x_test = torch.FloatTensor([[4,4]])
print(H(x_test).detach().item())
print(x_test.shape)
```
